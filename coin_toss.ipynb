```python
import numpy as np
h=0
t=0
mat = np.random.randint(1,20,(3,3))
print(mat)
toss_mat=mat.copy()
for i in range(0,3):
    for j in range(0,3):
        if toss_mat[i,j]%3==0:
            h+=1
            toss_mat[i,j]=0
        else:
            t+=1
            toss_mat[i,j]=1

print(toss_mat)
print("no of head=",h,"and no of tail=",t)
get_h=h/9
print('probability of getting head=',get_h)
print('probability of getting tail =',1-get_h)
```
